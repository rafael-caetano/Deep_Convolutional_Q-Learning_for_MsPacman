# Deep Convolutional Q-Learning for Pac-Man

This project implements and compares two deep reinforcement learning approaches for playing the Ms. Pac-Man game: the original Deep Q-Network (DQN) and the Dueling DQN. Using the `MsPacmanDeterministic-v0` environment from Gymnasium's Atari suite, we train agents to play the game based solely on raw pixel input.

Key features of this project include:
- Convolutional Neural Networks (CNNs) for processing game frames
- Implementation of both DQN and Dueling DQN architectures
- Frame preprocessing to optimize input for the neural networks
- Epsilon-greedy exploration strategy with decay
- Experience replay for improved learning stability
- Comparison of performance between DQN and Dueling DQN

The notebook demonstrates the entire process from environment setup and agent training to performance evaluation and visualization. We analyze the learning curves, compare the final performance of both architectures, and showcase the trained agents' gameplay through video recordings.
